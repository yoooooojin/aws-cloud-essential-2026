# 실습 6: 유명인 인식 (Celebrity Recognition)

## 학습 목표
- Amazon Rekognition의 `recognize_celebrities` API로 유명인을 식별합니다.
- 유명인 정보(이름, 관련 URL)와 인식 못한 얼굴을 처리합니다.

## API 개요
`recognize_celebrities`는 이미지에서 유명인을 탐지합니다.
- **반환값**:
  - `CelebrityFaces`: 인식된 유명인 목록
    - `Name`: 이름 / `MatchConfidence`: 매칭 신뢰도 / `Urls`: 관련 URL / `Face.BoundingBox`: 위치
  - `UnrecognizedFaces`: 인식되지 않은 얼굴 목록


```python
import boto3
import os
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# ✅ [제공 코드]
rekognition = boto3.client('rekognition', region_name='ap-northeast-2')
IMAGE_DIR = './images/'
image_filename = 'lab06.jpg'
image_path = os.path.join(IMAGE_DIR, image_filename)

def load_image_bytes(path):
    with open(path, 'rb') as f:
        return f.read()

image_bytes = load_image_bytes(image_path)
print(f"이미지 경로: {image_path}")
```

## ✏️ TODO 1: recognize_celebrities API 전체 작성

이 실습에서는 API 호출부터 결과 출력까지 직접 작성합니다.

**참고**: 이전 실습들의 패턴을 활용하세요!


```python
# ✏️ TODO 1: API를 호출하고 결과를 변수에 저장하세요
response = rekognition.recognize_celebrities( # ← API 메서드명
    Image={'Bytes': image_bytes}    # ← 이미지 바이트
)  # ← 전체 작성

celebrities  = response['CelebrityFaces']  # ← 'CelebrityFaces'
unrecognized = response['UnrecognizedFaces']  # ← 'UnrecognizedFaces'

print(f"인식된 유명인: {len(celebrities)}명")
print(f"인식 못한 얼굴: {len(unrecognized)}개")
```

## ✏️ TODO 2: 유명인 정보 출력


```python
# ✏️ TODO 2: 각 유명인의 이름과 신뢰도, URL을 출력하세요
print("\n인식된 유명인:")
print("-" * 50)
for celeb in celebrities:
    name = celeb['Name']        # ← 'Name'
    conf = celeb['MatchConfidence']        # ← 'MatchConfidence'
    urls = celeb.get('Urls', []) # ← 'Urls'
    print(f"  이름: {name}")
    print(f"  신뢰도: {conf:.1f}%")
    if urls:
        print(f"  참고 URL: {urls[0]}")
    print()
```

## ✏️ TODO 3: 유명인 얼굴에 이름 레이블 표시 (직접 작성)

이전 실습 패턴을 참고하여 인식된 유명인의 얼굴에 이름을 표시하는 코드를 처음부터 작성해보세요.


```python
# ✏️ TODO 3: 유명인 얼굴 시각화 코드를 처음부터 작성하세요



# 여기에 코드를 작성하세요
img = Image.open(image_path)
draw = ImageDraw.Draw(img)
w, h = img.size
for celeb in celebrities:
    box = celeb['Face']['BoundingBox'] # <- 힌트: 이전 실습(3, 4, 5)의 BoundingBox 처리 패턴을 참고하세요
    left   = box['Left'] * w
    top    = box['Top'] * h
    right  = left + box['Width'] * w
    bottom = top  + box['Height'] * h
    draw.rectangle([left, top, right, bottom], outline='gold', width=3)
    draw.text((left, top-20), celeb['Name'], fill='gold')
plt.figure(figsize=(10, 8))
plt.imshow(img)
plt.axis('off')
plt.title('Celebrity Recognition 결과')
plt.show()

```
